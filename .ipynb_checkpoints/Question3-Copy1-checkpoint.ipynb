```python
dict_name2avg_features = {}

for name in dict_name2imgfeature:
    average_embed = torch.mean(dict_name2imgfeature[name], dim=0)
    dict_name2avg_features[name] = average_embed
    
```

## Image to Topic Text Similarity (Average )




```python
topic_lst = [] 
img_feature_lst = []
key_lst = []

for key, img_features in dict_name2avg_features.items():
    #topic, country, income = key.split('_')
    key_lst.append(key)
    topic = key.split('_')[0]
    topic_lst.append(topic)
    img_feature_lst.append(img_features)
    #print(topic, income)
    #break
```

```python
topic_set = set(topic_lst)
topic_list = list(sorted(topic_set))
len(topic_list)
```

```python
with torch.no_grad():
    text_tokens = clip.tokenize([desc for desc in topic_list]).cuda()
    text_features = model.encode_text(text_tokens).float()
    text_features /= text_features.norm(dim=-1, keepdim=True)

text_features.shape
```

```python
img_feature_list= torch.stack([image_feature for image_feature in img_feature_lst])
```

```python
img_feature_list.shape
```

```python
similarity_img2topictext = text_features.cpu().numpy() @ img_feature_list.cpu().numpy().T # dot product of image and text features
similarity_img2topictext.shape
```

```python
similarity_img2topictext[1][8100]
```

```python
labels = []
for topic in topic_list:
    label = []
    for img in key_lst:
          label.append(topic + ' x ' + img)
    labels.append(label)
        
```

```python
topic_list
```

```python
len(labels[99])
```

```python
which_topic_a = 'kitchen sink'
which_imgs_a = 'kitchen sink x get water_Nigeria_poor'

topic_index_a = topic_list.index(which_topic_a)
 
image_index_a = labels[topic_index_a].index(which_imgs_a)

cos_sim_a = similarity_img2topictext[topic_index_a][image_index_a]

equivalents = []

for index1 in range(len(labels)):
    for index2 in range(len(labels[0])):
        if similarity_img2topictext[index1][index2] == cos_sim_a:
            equivalents.append([index1, index2])
            
            
            
for indexes in equivalents:
    print(labels[indexes[0]][indexes[1]] )
```
