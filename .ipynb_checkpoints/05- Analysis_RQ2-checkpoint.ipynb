```python
data = pd.read_csv(path + '/results_CLIP_no_sub.csv')
data.head()
```

```python
labels=["poor", "lower-middle", "upper-middle", "rich"]


data["quartile"] = pd.qcut(data["income"], q=[0, 0.25, 0.5, 0.75, 1],    
                         labels=labels)
```

```python
# list of subjective topics


subjective_topics = ['favorite home decorations', 'favourite item in kitchen', 
                    'favourite sports clubs', 'how the most loved item is used',
                    'icons', 'idols', 'latest furniture bought', ' looking over the shoulder',
                    'most loved item', 'most loved toy', 'most played songs on the radio',
                    'music idol', 'next big thing you are planning to buy',
                     'playing with most loved toy',
                    'thing i dream about having','things i wish i had',
                     'using most loved item', 'youth culture',
                    'what i wish i could buy']
```

```python
# map each topic to list of corresponding images
dict_topic2img = {}
count_ = []
for list_topics, image_path in zip(data['topics'], data['imageRelPath']):
    for topic in list_topics.split(","):
        topic = topic.lower()
        if topic not in subjective_topics: # remove subjectives
            
            if topic not in dict_topic2img:
                count_.append(topic)
                dict_topic2img[topic] = set()
            dict_topic2img[topic].add(image_path)

print(len(dict_topic2img))
```

```python
# map each image to corresponding topics, countries, income level ... TODO
dict_img2topic, dict_img2country, dict_img2incomelevel, dict_country2region = {}, {}, {}, {}
for list_topics, image_path, country, incomelevel, region in zip(data['topics'], data['imageRelPath'], data['country.name'], data['quartile'], data['region.id']):
    dict_img2topic[image_path] = list_topics
    dict_img2country[image_path] = country
    dict_img2incomelevel[image_path] = incomelevel
    dict_country2region[country] = region

print(len(dict_img2topic))
# print(dict_img2country)
```

```python
# query CLIP for each unique topic in set_topics
topic_count_dict = {}
for topic in dict_topic2img:
    images = dict_topic2img[topic]
    count = len(images)
    #print(topic, count)
```

```python
import pickle

file_to_read = open("dict_topic2score_old.pkl", "rb")

dict_topic2score2 = pickle.load(file_to_read)
```

```python
#compare dict_topic2score[:count] to dict_topic2img
from tqdm import tqdm
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# THRESHOLD = 50

all_images = list(data['imageRelPath'])
dict_topic2FNimg = {}
dict_topic2F1 = {}
for topic in tqdm(list(dict_topic2score2.keys())): #TODO: do for all topics[:1]
    if topic not in subjective_topics: 
        count_pos = len(dict_topic2img[topic])
#         if count_pos < THRESHOLD:
#             continue
#         print(f"topic '{topic}' has {count_pos} positives in a total of {len(dict_topic2score2[topic])}")

        GT_pos_images = dict_topic2img[topic]
        Pred_pos_images = [data[1] for data in dict_topic2score2[topic][:count_pos]]

        GT_labels = [1 if img in GT_pos_images else 0 for img in all_images]
        Pred_labels = [1 if img in Pred_pos_images else 0 for img in all_images]
        #print(Pred_labels)
        #print(Counter(GT_labels))
        #print(Counter(Pred_labels))

        accuracy = round(accuracy_score(GT_labels, Pred_labels) * 100, 2)
        recall = round(precision_score(GT_labels, Pred_labels) * 100, 2)
        precision = round(recall_score(GT_labels, Pred_labels) * 100, 2)
        f1 = round(f1_score(GT_labels, Pred_labels) * 100, 2)
        #print(f"Acc: {accuracy}, Recall: {recall}, Precision: {precision}, F1: {f1}")
        #print("---------")
        dict_topic2F1[topic] = f1

        # What are the forgotten items (FN) 
        images_FN = []
        for idx, img in enumerate(all_images):
            label_GT = GT_labels[idx]
            label_Pred = Pred_labels[idx]
            #FN
            if label_GT == 1 and label_Pred == 0:
                images_FN.append(img) #TODO: map img to country& income 
        #print(images_FN)
        dict_topic2FNimg[topic] = images_FN
    
```

```python
from collections import Counter

list_all_countriesFN = []
dict_topic2FNCountry = {}
dict_all_Countries_FN = {}
for topic in dict_topic2FNimg:
    list_imagesFN = dict_topic2FNimg[topic]
    list_countriesFN = [dict_img2country[img] for img in list_imagesFN]
    for country in list_countriesFN:
        list_all_countriesFN.append(country)
    #print(f"topic: {topic}; {Counter(list_countriesFN)}")
    dict_topic2FNCountry[topic] = Counter(list_countriesFN)
    
#print(f"ALL topics: {Counter(list_all_countriesFN)}")
dict_all_Countries_FN = Counter(list_all_countriesFN)
```

```python
# Country ground truth

list_all_countriesGT = []
dict_topic2GTCountry = {}
dict_all_Countries_GT = {}
for topic in dict_topic2img:
    list_imagesGT = dict_topic2img[topic]
    list_countriesGT = [dict_img2country[img] for img in list_imagesGT]
    for country in list_countriesGT:
        list_all_countriesGT.append(country)
    #print(f"topic: {topic}; {Counter(list_countriesFN)}")
    dict_topic2GTCountry[topic] = Counter(list_countriesGT)
    
#print(f"ALL topics: {Counter(list_all_countriesGT)}")
dict_all_Countries_GT = Counter(list_all_countriesGT)
```

```python
### Income level forgotten images


list_all_incomelevelsFN = []
dict_topic2FNIncomelevels = {}
dict_all_incomelevels_FN = {}
for topic in dict_topic2FNimg:
    list_imagesFN = dict_topic2FNimg[topic]
    list_incomelevelsFN = [dict_img2incomelevel[img] for img in list_imagesFN]
    for incomelevel in list_incomelevelsFN:
        list_all_incomelevelsFN.append(incomelevel)
    #print(f"topic: {topic}; {Counter(list_countriesFN)}")
    dict_topic2FNIncomelevels[topic] = Counter(list_incomelevelsFN)
    
#print(f"ALL topics: {Counter(list_all_incomelevelsFN)}")
dict_all_incomelevels_FN = Counter(list_all_incomelevelsFN)
```

```python
### Income level ground truth

list_all_incomelevelsGT = []
dict_topic2GTIncomelevels = {}
dict_all_incomelevels_GT = {}
for topic in dict_topic2img:
    list_imagesGT = dict_topic2img[topic]
    list_incomelevelsGT = [dict_img2incomelevel[img] for img in list_imagesGT]
    for incomelevel in list_incomelevelsGT:### Topics where Poor has higher Recall than other classes (Appendix)
        list_all_incomelevelsGT.append(incomelevel)
    #print(f"topic: {topic}; {Counter(list_countriesFN)}")
    dict_topic2GTIncomelevels[topic] = Counter(list_incomelevelsGT)
    
#print(f"ALL topics: {Counter(list_all_incomelevelsGT)}")
dict_all_incomelevels_GT = Counter(list_all_incomelevelsGT)
```

```python
# Forgotten images / total images per income level = Forget ratio
# Higher forget ratio indicates poorer CLIP performance for that income level
#dict_all_Countries_FN/dict_all_Countries_GT


Forget_ratio_percountry_allincome = Counter({key : dict_all_incomelevels_FN[key] / dict_all_incomelevels_GT[key] for key in dict_all_incomelevels_FN})
```

```python
Forget_ratio_perincomelevel_pertopic = {}
for topic in dict_topic2FNIncomelevels:
    FN_counter_ = dict_topic2FNIncomelevels[topic]
    GT_counter_ = dict_topic2GTIncomelevels[topic]
        
    temp_forget_ratio_ = Counter({key : FN_counter_[key] / GT_counter_[key] for key in FN_counter_ if sum(GT_counter_.values()) * 0.01 <= GT_counter_[key] })
    Forget_ratio_perincomelevel_pertopic[topic] = temp_forget_ratio_
```

```python
Usual_topics = []
for key in Forget_ratio_perincomelevel_pertopic.keys():
    poor = Forget_ratio_perincomelevel_pertopic[key]['poor']
    rich = Forget_ratio_perincomelevel_pertopic[key]['rich']
    low_mid = Forget_ratio_perincomelevel_pertopic[key]['lower-middle']
    up_mid = Forget_ratio_perincomelevel_pertopic[key]['upper-middle']
    if (poor > rich) and (poor > up_mid) and (poor > low_mid):
        Usual_topics.append(key)# Forget ratio is the false negative rate (FN/Ground truth) = 1 - Recall

Forget_ratio_percountry_alltopics = Counter({key : dict_all_Countries_FN[key] / dict_all_Countries_GT[key] for key in dict_all_Countries_FN})
```

```python
Forget_ratio_percountry_pertopic = {}
for topic in dict_topic2FNCountry:
    FN_counter = dict_topic2FNCountry[topic]
    GT_counter = dict_topic2GTCountry[topic]
        
    temp_forget_ratio = Counter({key : FN_counter[key] / GT_counter[key] for key in FN_counter if sum(GT_counter.values()) * 0.01 <= GT_counter[key] })
    Forget_ratio_percountry_pertopic[topic] = temp_forget_ratio
```

```python
# Recall

Recall_perincomelevel_pertopic = {}
for topic in dict_topic2FNIncomelevels:
    FN_counter_ = dict_topic2FNIncomelevels[topic]
    GT_counter_ = dict_topic2GTIncomelevels[topic]
        
    temp_recall_ratio_ = Counter({key : 1 - FN_counter_[key] / GT_counter_[key] for key in FN_counter_ if sum(GT_counter_.values()) * 0.01 <= GT_counter_[key] })
    Recall_perincomelevel_pertopic[topic] = temp_recall_ratio_
```
